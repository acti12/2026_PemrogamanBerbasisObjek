# Praktikum 1 - Tipe data Dasar

```python
# 1. Number (int, float, complex)
# Integer (int)
angka_int = 10
print(f"angka_int = {angka_int} -> {type(angka_int)}")

# Float
angka_float = 3.14
print(f"angka_float = {angka_float} -> {type(angka_float)}")

# Complex
angka_complex = 2 + 3j
print(f"angka_complex = {angka_complex} -> {type(angka_complex)}")

# 2. Boolean
is_active = True
print(f"is_active = {is_active} -> {type(is_active)}")

# 3. String
teks = "Hello, Python!"
print(f"teks = {teks} -> {type(teks)}")

# 4. List (Mutable: Bisa diubah)
daftar_angka = [1, 2, 3, 4, 5]
print(f"daftar_angka = {daftar_angka} -> {type(daftar_angka)}")

# 5. Tuple (Immutable: Tidak bisa diubah)
koordinat = (10, 20)
print(f"koordinat = {koordinat} -> {type(koordinat)}")

# 6. Dictionary (Pasangan Key-Value)
data_mahasiswa = {
    "nama": "Andi",
    "nim": "A11.2022.12345",
    "jurusan": "Teknik Informatika"
}
print(f"data_mahasiswa = {data_mahasiswa} -> {type(data_mahasiswa)}")

# 7. Set (Unik dan tidak terurut)
himpunan_angka = {1, 2, 3, 2, 1}
print(f"himpunan_angka = {himpunan_angka} -> {type(himpunan_angka)}")

# 8. Contoh penggunaan konversi tipe data
nilai_str = "100"
print(f"\nnilai_str = {nilai_str} -> {type(nilai_str)}")

nilai_int = int(nilai_str)  # Konversi string ke integer
print(f"nilai_int = {nilai_int} -> {type(nilai_int)}")
```

# Praktikum 2 - Variable Dan Operasi Dasar

```python
# -*- coding: utf-8 -*-

# 1. Pendeklarasian Variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print(f"Nama = {nama}")
print(f"Umur = {umur}")
print(f"Tinggi = {tinggi} cm")
print(f"Mahasiswa = {is_student}")

# Python tidak memerlukan deklarasi tipe data secara eksplisit.
# Tipe data variabel akan menyesuaikan nilai yang diberikan (Dynamic Typing).

# 2. Operasi Aritmetika
a = 10
b = 3

penjumlahan = a + b      # Tambah
pengurangan = a - b      # Kurang
perkalian = a * b        # Kali
pembagian = a / b        # Bagi (hasil float: 3.333...)
pembagian_bulat = a // b # Bagi (hasil dibulatkan ke bawah: 3)
modulus = a % b          # Sisa bagi (10/3 sisa 1)
pangkat = a ** b         # Pemangkatan (10 pangkat 3)

print("\nOPERASI ARITMETIKA")
print(f"a = {a}, b = {b}")
print(f"Penjumlahan = {penjumlahan}")
print(f"Pengurangan = {pengurangan}")
print(f"Perkalian = {perkalian}")
print(f"Pembagian = {pembagian}")
print(f"Pembagian Bulat = {pembagian_bulat}")
print(f"Modulus = {modulus}")
print(f"Pangkat = {pangkat}")

# 3. Operasi Perbandingan
# Menghasilkan nilai Boolean (True/False)
lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\nOPERASI PERBANDINGAN")
print(f"a > b = {lebih_besar}")
print(f"a < b = {kurang_dari}")
print(f"a == b = {sama_dengan}")
print(f"a != b = {tidak_sama}")
print(f"a >= b = {lebih_besar_sama}")
print(f"a <= b = {kurang_sama}")

# 4. Operasi Logika
# and, or, not
x = True
y = False

logika_and = x and y   # True jika keduanya True
logika_or = x or y     # True jika salah satu saja True
logika_not_x = not x   # Membalik nilai (True jadi False)

print("\nOPERASI LOGIKA")
print(f"x = {x}, y = {y}")
print(f"x and y = {logika_and}")
print(f"x or y = {logika_or}")
print(f"not x = {logika_not_x}")

# 5. Contoh penggunaan di dalam percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
else:
    print("\nKondisi tidak terpenuhi atau b <= 0.")
```

# Praktikum 3 - Percabangan

```python
# -*- coding: utf-8 -*-

# 1. IF Sederhana
# Program hanya mengeksekusi blok jika kondisinya True
nilai = 85
print("--- Contoh IF Sederhana ---")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")

# 2. IF-ELSE
# Jika kondisi True, eksekusi blok if; jika False, eksekusi blok else
umur = 17
print("--- Contoh IF-ELSE ---")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

# 3. IF-ELIF-ELSE
# Menangani banyak kondisi secara berurutan.
hari = "Rabu"
print("--- Contoh IF-ELIF-ELSE ---")
if hari == "Senin":
    print("Hari Senin - Saatnya kembali bekerja!")
elif hari == "Selasa":
    print("Hari Selasa - Jadwal rapat mingguan.")
elif hari == "Rabu":
    print("Hari Rabu - Ada diskon di beberapa toko.")
else:
    print("Hari lainnya - Atur jadwalmu dengan baik.\n")

# 4. IF Bersarang (Nested IF)
# Memeriksa sub-kondisi setelah kondisi pertama terpenuhi.
suhu = 35
print("--- Contoh IF Bersarang (Nested IF) ---")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.")
else:
    print("Cuaca sepertinya cukup sejuk.\n")

# 5. Menggabungkan Percabangan dengan Operasi Logika (AND/OR)
nilai_teori = 75
nilai_praktik = 80
print("--- Contoh IF dengan Operasi Logika ---")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")

# 6. Penggunaan If Ternary (Shorthand IF-ELSE)
angka = -5
print("--- Contoh If Ternary ---")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print(f"Angka = {angka} => {status}")
```

# Praktikum 4 - Perulangan

```python
# -*- coding: utf-8 -*-

# 1. FOR Loop dengan range()
print("1) FOR loop dengan range()")
for i in range(5):
    print(f"Perulangan ke-{i}")
# range(5) menghasilkan nilai 0, 1, 2, 3, dan 4

print()  # Pemisah output

# 2. FOR Loop untuk mengiterasi List
print("2) FOR loop mengiterasi list")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print(f"Buah: {item}")

print()

# 3. WHILE Loop
print("3) WHILE loop sederhana")
count = 0
while count < 5:
    print(f"count = {count}")
    count += 1  # Increment agar tidak terjadi infinite loop

print()

# 4. BREAK pada Loop
print("4) BREAK di dalam loop")
for i in range(10):
    if i == 3:
        print(f"Loop dihentikan secara paksa pada i = {i}")
        break  # Mengakhiri seluruh loop
    print(f"i = {i}")

print()

# 5. CONTINUE pada Loop
print("5) CONTINUE di dalam loop")
for i in range(5):
    if i == 2:
        print(f"--- Lewati i = {i} dengan continue ---")
        continue  # Langsung lompat ke iterasi berikutnya
    print(f"i = {i}")

print()

# 6. NESTED Loop (Loop Bersarang)
print("6) NESTED loop")
for i in range(3):           # Loop luar (outer loop)
    for j in range(2):       # Loop dalam (inner loop)
        print(f"i={i}, j={j}")

print()

# 7. Memanfaatkan ELSE pada Loop
print("7) ELSE pada loop for/while")
# Blok else dieksekusi HANYA jika loop selesai secara alami (tanpa break)
for x in range(3):
    print(f"x = {x}")
else:
    print("Loop for telah selesai tanpa interupsi break.\n")

y = 0
while y < 3:
    print(f"y = {y}")
    y += 1
else:
    print("Loop while telah selesai tanpa interupsi break.\n")

# 8. PASS sebagai placeholder
print("8) PASS (placeholder)")
for i in range(3):
    if i == 1:
        pass  # Tidak melakukan apa-apa, hanya agar tidak error saat blok kosong
    print(f"i = {i}")
```

# Tugas 2 - Program untuk mengecek suatu bilangan

```python
# =============================================
# Program 2: Cek Bilangan Genap, Ganjil, Prima
# =============================================

def cek_prima(n):
    """Fungsi untuk mengecek apakah bilangan n adalah prima."""
    if n < 2:
        return False          # 0 dan 1 bukan bilangan prima
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False      # Ada faktor pembagi → bukan prima
    return True               # Tidak ada faktor → prima


# Ambil input bilangan dari pengguna
angka = int(input("Masukkan sebuah bilangan: "))

# Cek genap atau ganjil
if angka % 2 == 0:
    jenis = "Genap"
else:
    jenis = "Ganjil"

# Cek prima
if cek_prima(angka):
    status_prima = "Ya, bilangan ini adalah bilangan PRIMA"
else:
    status_prima = "Bukan bilangan prima"

# Tampilkan hasil
print(f"\n===== HASIL PENGECEKAN =====")
print(f"Bilangan : {angka}")
print(f"Jenis    : {jenis}")
print(f"Prima    : {status_prima}")
```
