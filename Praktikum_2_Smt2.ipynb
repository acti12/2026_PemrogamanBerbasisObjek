# Praktikum 1 - Membuat dan Menggunakan Fungsi

```python
# 1. Fungsi Built-in
# Fungsi print() adalah contoh fungsi built-in untuk mencetak output
print("Ini adalah contoh fungsi built-in")

# Fungsi len() untuk menghitung panjang suatu objek
kata = "Pemrograman"
panjang_kata = len(kata)
print(f"Panjang kata '{kata}' adalah: {panjang_kata}")

# Fungsi max() untuk mencari nilai maksimum dari sebuah daftar
angka = [10, 5, 30, 40, 25]
nilai_max = max(angka)
print(f"Nilai maksimum dalam daftar {angka} adalah: {nilai_max}")

print("-" * 30)

# 2. Fungsi User-Defined (Dibuat oleh Pengguna)

# Fungsi dengan satu parameter
def cetak_kuadrat(angka):
    """Menghitung dan mencetak kuadrat dari angka."""
    hasil = angka ** 2
    print(f"Kuadrat dari {angka} adalah: {hasil}")

# Fungsi dengan beberapa parameter
def hitung_luas_persegi_panjang(panjang, lebar):
    """Mengembalikan nilai luas persegi panjang."""
    return panjang * lebar

# Fungsi dengan berbagai tipe parameter
def info_mahasiswa(nama, umur, ipk):
    """Mencetak informasi mahasiswa."""
    print(f"Nama: {nama}, Umur: {umur}, IPK: {ipk}")

# Fungsi tanpa return value (Prosedur)
def sapa_pengguna(nama):
    print(f"Halo, {nama}! Selamat datang di dunia Python.")

# Fungsi dengan return value
def hitung_keliling_persegi(sisi):
    return 4 * sisi

# --- Pemanggilan Fungsi ---

# Memanggil fungsi satu parameter
cetak_kuadrat(5)

# Memanggil fungsi beberapa parameter dan menyimpan hasilnya
luas = hitung_luas_persegi_panjang(10, 5)
print(f"Luas persegi panjang: {luas}")

# Memanggil fungsi dengan berbagai tipe data
info_mahasiswa("Budi", 22, 3.8)

# Memanggil fungsi tanpa return value
sapa_pengguna("Andi")

# Memanggil fungsi dengan return value
keliling = hitung_keliling_persegi(5)
print(f"Keliling persegi dengan sisi 5 adalah: {keliling}")
```

# Praktikum 3 - Membuat Kelas Sederhana

```python
# --- KELAS BUKU ---
class Buku:
    def __init__(self, judul, pengarang, tahun_terbit):
        self.judul = judul
        self.pengarang = pengarang
        self.tahun_terbit = tahun_terbit
        self.status = "Tersedia"  # Status default

    def tampilkan_info(self):
        print(f"Judul: {self.judul}")
        print(f"Pengarang: {self.pengarang}")
        print(f"Tahun Terbit: {self.tahun_terbit}")
        print(f"Status: {self.status}")

    def pinjam(self):
        if self.status == "Tersedia":
            self.status = "Dipinjam"
            print(f"Buku '{self.judul}' telah dipinjam.")
        else:
            print(f"Buku '{self.judul}' sedang dipinjam.")

    def kembalikan(self):
        if self.status == "Dipinjam":
            self.status = "Tersedia"
            print(f"Buku '{self.judul}' telah dikembalikan.")
        else:
            print(f"Buku '{self.judul}' tidak sedang dipinjam.")

# Membuat objek dari kelas Buku
buku1 = Buku("Pemrograman Python", "John Doe", 2021)
buku2 = Buku("Data Science untuk Pemula", "Jane Smith", 2020)

print("--- Info Buku ---")
buku1.tampilkan_info()
buku2.pinjam()
buku2.kembalikan()
buku1.pinjam()

# --- KELAS MAHASISWA ---
class Mahasiswa:
    def __init__(self, nama, nim, umur):
        self.nama = nama
        self.nim = nim
        self.umur = umur

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"Umur: {self.umur} tahun")

    def ubah_umur(self, umur_baru):
        self.umur = umur_baru

# Membuat objek dari kelas Mahasiswa
mahasiswa1 = Mahasiswa("Andi", "12345", 20)
mahasiswa2 = Mahasiswa("Budi", "67890", 22)

print("\n--- Info Mahasiswa ---")
mahasiswa1.tampilkan_info()
print()
mahasiswa2.tampilkan_info()

# Mengubah umur mahasiswa1
mahasiswa1.ubah_umur(21)
print("\nSetelah mengubah umur mahasiswa1:")
mahasiswa1.tampilkan_info()
```

# Praktikum 4 - Membuat dan Menggunakan Method dalam Kelas

```python
# Nama File: mobil.py

class Mobil:
    def __init__(self, merk, warna, tahun, harga):
        # Atribut: Variabel yang menyimpan data di dalam objek
        self.merk = merk
        self.warna = warna
        self.tahun = tahun
        self.harga = harga

    # 1. Metode tanpa return value (Prosedur)
    def tampilkan_info(self):
        print(f"Mobil {self.merk} berwarna {self.warna}, tahun {self.tahun}, harga: Rp {self.harga}")

    # 2. Metode dengan satu parameter
    def diskon(self, persen_diskon):
        diskon_harga = self.harga * (persen_diskon / 100)
        harga_setelah_diskon = self.harga - diskon_harga
        print(f"Harga {self.merk} setelah diskon {persen_diskon}%: Rp {harga_setelah_diskon}")

    # 3. Metode dengan return value
    def hitung_usia(self, tahun_sekarang):
        usia = tahun_sekarang - self.tahun
        return usia

    # 4. Metode dengan beberapa parameter
    def perbarui_harga(self, harga_baru, tahun_baru):
        self.harga = harga_baru
        self.tahun = tahun_baru
        print(f"Data {self.merk} diperbarui: Rp {self.harga} (Tahun {self.tahun})")


# --- Contoh Penggunaan ---

# Membuat objek (Instansiasi)
mobil1 = Mobil("Toyota", "Hitam", 2015, 300000000)
mobil2 = Mobil("Honda", "Merah", 2018, 250000000)

# Memanggil metode info
mobil1.tampilkan_info()
mobil2.tampilkan_info()

# Memanggil metode diskon
mobil1.diskon(10)
mobil2.diskon(15)

# Memanggil metode dengan return value
# Hasil return disimpan ke dalam variabel
usia_mobil1 = mobil1.hitung_usia(2025)
usia_mobil2 = mobil2.hitung_usia(2025)

print(f"Usia {mobil1.merk}: {usia_mobil1} tahun")
print(f"Usia {mobil2.merk}: {usia_mobil2} tahun")

# Memperbarui data
mobil1.perbarui_harga(280000000, 2022)
```

# Tugas 1 - Program Sederhana Jadwal Kelas

```python
# ============================================================
# Nama File  : main_jadwal.py
# Deskripsi  : Program utama Jadwal Kelas yang menggabungkan
#              konsep Modularisasi (Praktikum 02) dan
#              Kelas Sederhana (Praktikum 03).
# ============================================================

# 1. Mengimpor modul jadwal_kelas yang telah dibuat
import jadwal_kelas


# ============================================================
# BAGIAN KELAS (Praktikum 03)
# Mendefinisikan blueprint objek "Jadwal"
# ============================================================

class Jadwal:
    """Kelas untuk merepresentasikan satu jadwal mata kuliah."""

    def __init__(self, mata_kuliah, dosen, hari, jam_mulai, jam_selesai, ruangan, sks):
        # Atribut objek (data yang dimiliki setiap jadwal)
        self.mata_kuliah = mata_kuliah
        self.dosen       = dosen
        self.hari        = hari
        self.jam_mulai   = jam_mulai
        self.jam_selesai = jam_selesai
        self.ruangan     = ruangan
        self.sks         = sks
        self.status      = "Aktif"   # Status default saat dibuat

    def tampilkan_info(self):
        """Menampilkan seluruh informasi jadwal secara terformat."""
        print(f"  Mata Kuliah : {self.mata_kuliah}")
        print(f"  Dosen       : {self.dosen}")
        print(f"  Hari        : {self.hari}")
        print(f"  Waktu       : {self.jam_mulai} - {self.jam_selesai}")
        print(f"  Ruangan     : {self.ruangan}")
        print(f"  SKS         : {self.sks}")
        print(f"  Status      : {self.status}")

    def batalkan(self):
        """Mengubah status jadwal menjadi Dibatalkan."""
        if self.status == "Aktif":
            self.status = "Dibatalkan"
            print(f"  [!] Jadwal '{self.mata_kuliah}' telah DIBATALKAN.")
        else:
            print(f"  [i] Jadwal '{self.mata_kuliah}' sudah berstatus '{self.status}'.")

    def aktifkan_kembali(self):
        """Mengubah status jadwal kembali menjadi Aktif."""
        if self.status == "Dibatalkan":
            self.status = "Aktif"
            print(f"  [v] Jadwal '{self.mata_kuliah}' telah DIAKTIFKAN kembali.")
        else:
            print(f"  [i] Jadwal '{self.mata_kuliah}' sudah berstatus '{self.status}'.")

    def ganti_ruangan(self, ruangan_baru):
        """Memperbarui ruangan untuk jadwal ini."""
        ruangan_lama    = self.ruangan
        self.ruangan    = ruangan_baru
        print(f"  [>] Ruangan '{self.mata_kuliah}' diubah: {ruangan_lama} -> {ruangan_baru}")


# ============================================================
# BAGIAN PROGRAM UTAMA (main)
# Menggunakan kelas Jadwal + fungsi dari modul jadwal_kelas
# ============================================================

def main():

    # --- A. Membuat objek-objek Jadwal ---
    j1 = Jadwal("Pemrograman Lanjut",   "Dr. Rina Sari",   "Senin",   "08:00", "10:30", "Lab A1", 3)
    j2 = Jadwal("Basis Data",           "Bpk. Hendra",     "Selasa",  "10:00", "12:30", "R. 201",  3)
    j3 = Jadwal("Jaringan Komputer",    "Ibu. Dewi",       "Rabu",    "13:00", "15:30", "R. 305",  3)
    j4 = Jadwal("Matematika Diskrit",   "Dr. Ahmad Fauzi", "Senin",   "13:00", "14:40", "R. 102",  2)
    j5 = Jadwal("Kecerdasan Buatan",    "Dr. Lisa Putri",  "Kamis",   "08:00", "10:30", "Lab B2",  3)

    # Menyimpan semua jadwal dalam sebuah list
    semua_jadwal = [j1, j2, j3, j4, j5]


    # --- B. Menampilkan semua jadwal (fungsi modul + metode kelas) ---
    jadwal_kelas.format_header("SISTEM INFORMASI JADWAL KELAS")

    for i, jadwal in enumerate(semua_jadwal, start=1):
        print(f"\n  [Jadwal #{i}]")
        jadwal.tampilkan_info()    # Memanggil metode dari objek Jadwal
        jadwal_kelas.format_pemisah()


    # --- C. Menggunakan fungsi modul: hitung total SKS ---
    print()
    jadwal_kelas.format_header("TOTAL SKS")
    total_sks = jadwal_kelas.hitung_total_sks(semua_jadwal)
    print(f"  Total SKS seluruh mata kuliah: {total_sks} SKS")


    # --- D. Menggunakan fungsi modul: filter jadwal per hari ---
    print()
    jadwal_kelas.format_header("JADWAL HARI SENIN")

    hari_dicari = "Senin"
    # Validasi hari menggunakan fungsi dari modul
    if jadwal_kelas.validasi_hari(hari_dicari):
        jadwal_senin = jadwal_kelas.filter_jadwal_per_hari(semua_jadwal, hari_dicari)
        if jadwal_senin:
            for j in jadwal_senin:
                print()
                j.tampilkan_info()
                jadwal_kelas.format_pemisah()
        else:
            print(f"  Tidak ada jadwal pada hari {hari_dicari}.")
    else:
        print(f"  '{hari_dicari}' bukan hari yang valid.")


    # --- E. Demonstrasi perubahan status & ruangan (metode kelas) ---
    print()
    jadwal_kelas.format_header("PERUBAHAN JADWAL")

    j2.batalkan()                       # Batalkan Basis Data
    j2.batalkan()                       # Coba batalkan lagi (sudah dibatalkan)
    j2.aktifkan_kembali()               # Aktifkan kembali
    j3.ganti_ruangan("Lab C3")          # Ganti ruangan Jaringan Komputer

    print()
    print("  [Info terbaru Jaringan Komputer]")
    j3.tampilkan_info()


# --- Titik masuk program ---
if __name__ == "__main__":
    main()
```
